```python
import json
import pandas as pd

# read JSON data from file
with open('Car_info.json', 'r') as f:
    json_data = json.load(f)

# create dataframe from JSON data
car_df = pd.DataFrame(json_data)
car_df
```

```python
# group car_df by "model" column
groups = car_df.groupby('model')

# create a new dataframe Q2_df from the grouped data
Q2_df = groups.mean().reset_index()
Q2_df
```

```python
# group car_df by "make" column and compute summary statistics for "price" column
Q3_df = car_df.groupby('make')['price'].agg(['mean', 'min', 'max'])

# reset the index to make "make" a regular column
Q3_df = Q3_df.reset_index()

# rename the columns for clarity
Q3_df = Q3_df.rename(columns={'mean': 'mean_price', 'min': 'min_price', 'max': 'max_price'})
Q3_df
```

```python
# group car_df by "year" and "used" columns
groups = car_df.groupby(['year', 'used'])

# create a new dataframe Q4_df from the grouped data
Q4_df = groups.mean().reset_index()
Q4_df
```

```python
# group car_df by "color" column
groups = car_df.groupby('color')

# get the group with the name "red"
Q5_df = groups.get_group('Race Red')
Q5_df
```

```python
# create a new column "discount" in car_df that contains the amount of the 20% discount from the "price" column
discount_df = car_df.assign(discount = 0.2 * car_df['price'])
discount_df
```

```python
# concatenate car_df and discount_df along the columns axis
car_info_new = pd.concat([car_df, discount_df['discount']], axis=1)
car_info_new
```
